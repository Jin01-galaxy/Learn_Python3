```python
hp = 100
mp = 100
print('hp:',hp,'mp:',mp)
```

```python
def move_by_car():
    global hp 
    hp = hp + 70
    global mp 
    mp = mp - 100
    print(hp,mp)
move_by_car()
```

```python
def move_by_train():
    global hp 
    hp = hp + 30
    global mp 
    mp = mp - 50
    print(hp,mp)
move_by_train()
```
